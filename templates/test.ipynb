```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Niels\Desktop\Programming\Manifesting pasado programming\Datasets\Predictive Mainte\datasets initial.csv")
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df['DATE RETURNED'] = pd.to_datetime(df['DATE RETURNED'])
```

```python
df['TIME BORROWED'] = pd.to_datetime(df['TIME BORROWED'], format='%I:%M %p').dt.hour
df['TIME RETURNED'] = pd.to_datetime(df['TIME RETURNED'], format='%I:%M %p').dt.hour
```

```python
grouped = df.groupby('EQUIPMENT NAME')
```

```python
# Calculate mean and median for 'QUANTITY', 'DAYS DURATION', 'TIME DURATION', and 'TOTAL DURATION'
mean_per_equipment = grouped[['QUANTITY', 'DAYS DURATION', 'TIME DURATION', 'TOTAL DURATION']].mean()
median_per_equipment = grouped[['QUANTITY', 'DAYS DURATION', 'TIME DURATION', 'TOTAL DURATION']].median()
```

```python
print("Mean per equipment:\n", mean_per_equipment)
print("\nMedian per equipment:\n", median_per_equipment)
```

```python
mode_per_equipment = grouped[['QUANTITY', 'DAYS DURATION', 'TIME DURATION', 'TOTAL DURATION']].agg(pd.Series.mode)
std_per_equipment = grouped[['QUANTITY', 'DAYS DURATION', 'TIME DURATION', 'TOTAL DURATION']].std()
var_per_equipment = grouped[['QUANTITY', 'DAYS DURATION', 'TIME DURATION', 'TOTAL DURATION']].var()
min_per_equipment = grouped[['QUANTITY', 'DAYS DURATION', 'TIME DURATION', 'TOTAL DURATION']].min()
max_per_equipment = grouped[['QUANTITY', 'DAYS DURATION', 'TIME DURATION', 'TOTAL DURATION']].max()
```

```python
print("\nMode per equipment:\n", mode_per_equipment)
print("\nStandard deviation per equipment:\n", std_per_equipment)
print("\nVariance per equipment:\n", var_per_equipment)
print("\nMin per equipment:\n", min_per_equipment)
print("\nMax per equipment:\n", max_per_equipment)
```
